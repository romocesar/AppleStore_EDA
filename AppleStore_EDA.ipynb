```python
#Exploratory Data Analysis
#Count of apps by Track_name
#Count of apps by user_rating < 1, user_rating < 2, user_rating < 3, user_rating < 4,
#...user_rating < 5

user_rat_1 = appl_data[appl_data.user_rating < 1].id.count()
user_rat_2 = appl_data[(appl_data.user_rating >= 1) & (appl_data.user_rating < 2)].id.count()
user_rat_3 = appl_data[(appl_data.user_rating >= 2) & (appl_data.user_rating < 3)].id.count()
user_rat_4 = appl_data[(appl_data.user_rating >= 3) & (appl_data.user_rating < 4)].id.count()
user_rat_5 = appl_data[(appl_data.user_rating >= 4) & (appl_data.user_rating <= 5)].id.count()
print('User Rating 0-1:', user_rat_1)
print('User Rating 1-2:', user_rat_2)
print('User Rating 2-3:', user_rat_3)
print('User Rating 3-4:', user_rat_4)
print('User Rating 4-5:', user_rat_5)
tot_rat = (user_rat_1 + user_rat_2 + user_rat_3 + user_rat_4 + user_rat_5)
print('Total Ratings: ', tot_rat)
```

```python
#User Rating By percentage 
user_rat_1_per = user_rat_1/tot_rat*100
user_rat_2_per = user_rat_2/tot_rat*100
user_rat_3_per = user_rat_3/tot_rat*100
user_rat_4_per = user_rat_4/tot_rat*100
user_rat_5_per = user_rat_5/tot_rat*100

print('% User Rating 0-1:', user_rat_1_per)
print('% User Rating 1-2:', user_rat_2_per)
print('% User Rating 2-3:', user_rat_3_per)
print('% User Rating 3-4:', user_rat_4_per)
print('% User Rating 4-5:', user_rat_5_per)
```

```python
#Plot data
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline

#sns.set_style('dark')
ratings = [user_rat_1, user_rat_2, user_rat_3, user_rat_4, user_rat_5]
label_ratings = ['0-1 stars', '1-2 stars', '2-3 stars', '3-4 stars', '4-5 stars']
plt.pie(ratings, labels = label_ratings )
plt.title('Apple Store User Ratings')
plt.savefig('/Users/cesarromo/Desktop/AppleStore_ratings.jpg')
plt.show()

```
